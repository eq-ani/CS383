## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows
duplicate_numerical = numerical_df[numerical_df.duplicated()]
print(f"Total duplicate rows in numerical dataset: {len(duplicate_numerical)}")
print(duplicate_numerical)
# Drop these duplicate rows
numerical_df.drop_duplicates(inplace=True)
# Repeat the same for categorical dataset. Print the duplicate rows and drop them
duplicate_categorical = categorical_df[categorical_df.duplicated()]
print(f"Total duplicate rows in categorical dataset: {len(duplicate_categorical)}")
print(duplicate_categorical)
# Drop these duplicate rows
categorical_df.drop_duplicates(inplace=True)

````

## Part 3.2: Combine two differents datasets

A good skill to have is to know how to combine 2 different datasets.

Are all the unique ids are present in both datasets? Why do you think so? If not, what do the rows that are missing from one of the datasets look like in the combined table?

Answer: No, not all Unique Id's are present in both datasets. The rows that are missing from the numerical dataset would simply present in the merged table as "NaN" for all values that were supposed to be numerical. For example: 47 208034 NaN NaN NaN NaN NaN NaN NaN NaN NaN NaN NaN NaN NaN NaN NaN notpresent notpresent yes yes yes good yes no ckd, where ALL NaN values are from the numerical dataset, where unique id 208034 is not present. 

````python

# Merge the two given numerical and categorical datasets based on their unique_ID.
if 'unique_id' in categorical_df.columns and 'unique_id' in numerical_df.columns:
    merged_df = pd.merge(numerical_df, categorical_df, on="unique_id", how="outer")
else:
    raise ValueError("unique_id column is missing from one of the datasets")
#Print the combined dataset
print(merged_df.head()) 
print(merged_df.tail()) 

````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python

# Count rows with at least one missing value
rows_with_missing = merged_df.shape[0] - merged_df.dropna().shape[0]
missing_percentage = (rows_with_missing / merged_df.shape[0]) * 100
# Print before dropping missing values
print(f"Percentage of rows with missing values: {missing_percentage:.2f}%")
print("Before dropping missing values:")
print(merged_df.head()) 
print(merged_df.tail()) 
# Drop rows with missing values
merged_df.dropna(inplace=True)
# Print the dataset after removing missing values
print("After dropping missing values:")
merged_df.reset_index(drop=True, inplace=True)
print(merged_df.head()) 
print(merged_df.tail()) 
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting
if 'Target' in merged_df.columns:
    merged_df.sort_values(by='Target', inplace=True)
    merged_df.reset_index(drop=True, inplace=True)
else:
    raise ValueError("Target column is missing from the dataset")
# Print the dataset
print(merged_df.head()) 
print(merged_df.tail()) 
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['Red Blood Cells', 'Pus Cell', 'Pus Cell Clumps', 'Bacteria', 'Hypertension', 'Diabetes Mellitus', 'Coronary Artery Disease', 'Appetite', 'Pedal Edema', 'Anemia', 'Target']
merged_df = pd.get_dummies(merged_df, columns=categorical_columns, drop_first=False).astype(int)
# Print the dataset
print(merged_df.head()) 
print(merged_df.tail()) 
````

In the example we went through above, another solution is to have a single column for the binary variable. In the downstream modeling would this be equivalent? What effect would this have if the categorical variable could take more than 2 values? For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

Answer: Yes, it would be equivalent and actually take less memory since you can imply one column based on the other, since they can't be both. For example, if we had rbc_normal (true), then you would know that for that row, rbc_abnormal (false) must follow. This wouldn't be possible for something with more than 2 values because you can't imply what value the other one is given a false value in the column. An example with the condiments is if we had one row called c_ketchup, (meaning condiment ketchup) as false, then you wouldn't know whether it was mustard, mayo, ranch, or otherwise. 

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['Age', 'Blood Pressure', 'Blood Glucose Random', 'Blood Urea', 'Serum Creatinine', 'Sodium', 'Potassium', 'Hemoglobin', 'Packed Cell Volume', 'White Blood Cell Count', 'Red Blood Cell Count']
for col in numerical_columns:
    mean = merged_df[col].mean()
    std = merged_df[col].std()
    merged_df = merged_df[(merged_df[col] > mean - 3 * std) & (merged_df[col] < mean + 3 * std)]
# Print the dataset
merged_df.reset_index(drop=True, inplace=True)
print(merged_df.head()) 
print(merged_df.tail()) 
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
scaler = StandardScaler()
merged_df[numerical_columns] = scaler.fit_transform(merged_df[numerical_columns])
# Print the dataset
print(merged_df.head()) 
print(merged_df.tail()) 
````

## Part 3.8: Remove Unnecessary columns

Are there any columns in this dataset which are not appropriate for modeling and predictions? Which column(s)? Justify their exclusion and remove them

Answer: At this point in the calculations, since each row is considered a unique value and we aren't using these values to track patients, there no real need to have unique id's, and instead we can just opt to use the index values provided. Having the ID's are just needless clutter. 

````python
#Remove that column
unnecessary_columns = ['unique_id'] if 'unique_id' in merged_df.columns else []
merged_df.drop(columns=unnecessary_columns, inplace=True)
# Print the dataset
print(merged_df.head()) 
print(merged_df.tail()) 
````
